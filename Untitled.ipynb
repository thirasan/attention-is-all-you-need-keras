```python
from pythainlp.tokenize import word_tokenize
from tqdm import tqdm
```

```python
def FreqDict2List(dt):
	return sorted(dt.items(), key=lambda d:d[-1], reverse=True)
```

```python
def LoadCSV(fn):
    ret = []
    with open(fn, encoding='utf-8') as fin:
        for line in fin:
            lln = line.rstrip('\r\n').split('\t')
            ret.append(lln)
    return ret
```

```python
def LoadCSV(fn):
    ret = []
    with open(fn, encoding='utf8') as fin:
        sentence_pairs = []
        for line in fin:
            sentence_pairs.append(line.replace('\n', ''))
            if(len(sentence_pairs) == 2):
                ret.append(sentence_pairs)
                sentence_pairs = sentence_pairs[1:]
    return ret
```

```python
def LoadCSVg(fn):
    with open(fn, encoding='utf-8') as fin:
        sentence_pairs = []
        for line in fin:
            sentence_pairs.append(line.replace('\n', ''))
            if(len(sentence_pairs) == 2):
                yield sentence_pairs
                sentence_pairs = sentence_pairs[1:]
```

```python
class TokenList:
	def __init__(self, token_list):
		self.id2t = ['<PAD>', '<UNK>', '<S>', '</S>'] + token_list
		self.t2id = {v:k for k,v in enumerate(self.id2t)}
	def id(self, x):	return self.t2id.get(x, 1)
	def token(self, x):	return self.id2t[x]
	def num(self):		return len(self.id2t)
	def startid(self):  return 2
	def endid(self):    return 3
```

```python
def LoadCSVg(fn):
    with open(fn, encoding='utf-8') as fin:
        for line in fin:
            lln = line.rstrip('\r\n').split('\t')
            yield lln
```

```python
data = LoadCSV('mini.txt')
token_dict = {}
for pairs in tqdm(data):
    for token in word_tokenize(pairs[0],engine='deepcut') + word_tokenize(pairs[1],engine='deepcut'):
        if token not in token_dict:
            token_dict[token] = len(token_dict)
```

```python
data = LoadCSVg('mini.txt')
Xs = [[], []]
for ss in data:
    for seq, xs in zip(ss, Xs):
        xs.append(list(word_tokenize(seq,engine='deepcut')))
```
